# Hypothesis and Cost
### Hypothesis 
$$ H(x) = Wx + b $$


### Cost
$$ cost(W)=\frac { 1 }{ m } \sum _{i=1}^{m}{ { (W{ x }^{ i }-y^{ i } })^{ 2 } }  $$

```
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

W = tf.Variable(2.0)
b = tf.Variable(0.5)

hypothesis = W * x_data + b
```

```
W, b, hypothesis
```

```
plt.figure(figsize=(5,5))
plt.plot(x_data, hypothesis, 'r-')
plt.plot(x_data, y_data, 'o')
plt.grid()
plt.ylim(0, 10) # plt.ylim(최소, 최대)
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('Dataset')
plt.show()
```

## Cost

```
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

```
with tf.GradientTape() as tape:
    hypothesis = W * x_data + b
    cost = tf.reduce_mean(tf.square(hypothesis - y_data))

W_grad, b_grad = tape.gradient(cost, [W, b])
W_grad.numpy(), b_grad.numpy()
```

## 파라미터 업데이트

```
learning_rate = 0.01

W.assign_sub(learning_rate * W_grad)
b.assign_sub(learning_rate * b_grad)

print(W.numpy(), b.numpy())
```

```
plt.figure(figsize=(5,5))
plt.plot(x_data, hypothesis, 'r-')
plt.plot(x_data, y_data, 'o')
plt.grid()
plt.ylim(0, 10) # plt.ylim(최소, 최대)
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('Dataset')
plt.show()
```

## 여러 번 반복

```
W = tf.Variable(2.9)
b = tf.Variable(0.5)

for i in range(100):
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    W_grad, b_grad = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
    if i % 10 == 0:
      print("{:5}|{:10.4f}|{:10.4f}|{:10.6f}".format(i, W.numpy(), b.numpy(), cost))

plt.figure(figsize=(5,5))
plt.plot(x_data, hypothesis, 'r-')
plt.plot(x_data, y_data, 'o')
plt.grid()
plt.ylim(0, 10) # plt.ylim(최소, 최대)
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('Dataset')
plt.show()
```

# predict

```
print(W * 5 + b)
print(W * 2.5 + b)
```

## 전체 코드

```
import tensorflow as tf
import numpy as np

# Data
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

# W, b initialize
W = tf.Variable(2.9)
b = tf.Variable(0.5)

# W, b update
for i in range(100):
    # Gradient descent
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    W_grad, b_grad = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
    if i % 10 == 0:
      print("{:5}|{:10.4f}|{:10.4f}|{:10.6f}".format(i, W.numpy(), b.numpy(), cost))

print()

# predict
print(W * 5 + b)
print(W * 2.5 + b)
```
